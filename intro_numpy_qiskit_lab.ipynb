## Section 0: Python + NumPy + Matplotlib quickstart
This is a short warm-up to remove friction before we use quantum concepts.

```python
# Basic Python assignments and types
a = 3
b = 2.5
c = 1 + 2j

print('a type:', type(a))
print('b type:', type(b))
print('c type:', type(c))

# Lists and slicing
xs = [0, 1, 2, 3, 4, 5]
print('xs[1:4] =', xs[1:4])

# A simple function
def energy(m, c):
    return m * c**2

print('E =', energy(2, 3))
```

```python
# NumPy arrays: vectors and matrices
v = np.array([1, 2, 3], dtype=float)
w = np.array([3, 0, -1], dtype=float)
A = np.array([[1, 2], [3, 4]], dtype=float)

print('v + w =', v + w)
print('dot(v, w) =', np.dot(v, w))
print('A @ [1, 1] =', A @ np.array([1, 1]))
```

```python
# Complex numbers and Euler's formula
theta = np.pi / 3
z = np.exp(1j * theta)

print('z =', z)
print('|z| =', np.abs(z))
print('conj(z) =', np.conj(z))
```

```python
# Matplotlib basics: plot sine and cosine
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(6, 3))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.title('Sine and Cosine')
plt.show()
```

```python
# Visualizing complex numbers in the plane
points = [1 + 0j, 0.5 + 0.5j, -0.5 + 0.8j, -1 - 0.2j]

plt.figure(figsize=(4, 4))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)
plt.scatter([p.real for p in points], [p.imag for p in points])
for p in points:
    plt.text(p.real + 0.03, p.imag + 0.03, f'{p.real:.1f}+{p.imag:.1f}j')
plt.xlabel('Re')
plt.ylabel('Im')
plt.title('Complex Plane')
plt.gca().set_aspect('equal', 'box')
plt.show()
```

## Section 1: State vectors and measurement
A single-qubit state is a 2D complex vector.
We write |psi> = alpha|0> + beta|1>, with |alpha|^2 + |beta|^2 = 1.

```python
# Basis states |0> and |1>
ket0 = np.array([[1], [0]], dtype=complex)
ket1 = np.array([[0], [1]], dtype=complex)

# Build a state |psi>
alpha = 1/np.sqrt(3)
beta = np.sqrt(2/3)
psi = alpha * ket0 + beta * ket1

# Normalization check
norm = np.vdot(psi, psi)
print('psi =
', psi)
print('norm =', norm)
```

```python
# Measurement probabilities for |0> and |1>
p0 = np.abs(alpha)**2
p1 = np.abs(beta)**2

print('P(0) =', p0)
print('P(1) =', p1)

plt.figure(figsize=(4, 3))
plt.bar(['0', '1'], [p0, p1])
plt.ylim(0, 1)
plt.title('Measurement Probabilities')
plt.show()
```

## Section 2: Gates as matrices
Quantum gates are unitary matrices. Unitarity means U^dagger U = I.

```python
X = np.array([[0, 1], [1, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)

# Apply H to |0> (creates |+>)
plus = H @ ket0
print('|+> =
', plus)

# Apply Z to |+>
minus = Z @ plus
print('Z|+> =
', minus)

# Unitarity check
I = np.eye(2)
print('H unitary:', np.allclose(H.conj().T @ H, I))
```

## Section 3: First Qiskit circuit
We will build and simulate a single-qubit circuit.

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)

qc.draw('text')
```

```python
# Visual circuit (requires matplotlib)
qc.draw('mpl')
```

```python
# Run on simulator
result = SIM_BACKEND.run(qc, shots=1024).result()
counts = result.get_counts()
print(counts)
```

## Section 4: Statevector and Bloch sphere visualization
Qiskit can give the exact statevector (before measurement).

```python
from qiskit.quantum_info import Statevector

sv = Statevector.from_instruction(QuantumCircuit(1))
print('Initial state:', sv)

qc_state = QuantumCircuit(1)
qc_state.h(0)
sv_after = Statevector.from_instruction(qc_state)
print('After H:', sv_after)
```

```python
from qiskit.visualization import plot_bloch_multivector

plot_bloch_multivector(sv_after)
```
